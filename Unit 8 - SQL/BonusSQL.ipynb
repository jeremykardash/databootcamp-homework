```python
#Import dependencies
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create engine and inspector
engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/employees")
inspector = inspect(engine)
```

```python
#Create connection
cnx = engine.connect()
```

```python
#Read SQL database with pandas
import pandas as pd
salaries = pd.read_sql("SELECT * FROM salaries", cnx, parse_dates=['birth_date', 'hire_date'])
title = pd.read_sql("SELECT * from titles", cnx)
employee = pd.read_sql("SELECT * from employees", cnx, parse_dates=['birth_date', 'hire_date'])
```

```python
#Create histogram for most common salary ranges
x = salaries["salary"]
num_bins = 6
n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
plt.title("Histogram of Salries")
plt.xlabel("Salary Bins")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

```python
#Merge dataframes
merged = pd.merge(employee, salaries, on="emp_no")
df = merged.groupby(merged.emp_title_id).mean()
title_df = pd.merge(df, title, left_on = "emp_title_id", right_on = "id")
```

```python
#Create variables for bar chart
xticks = title_df["title"]
salary_group = title_df["salary"]
x_axis = np.arange(len(salary_group))

#Plot bar
plt.bar(x_axis, salary_group, color="r", width=0.5, alpha=0.5)
tick_location = [value for value in x_axis]
plt.xticks(tick_location, xticks, rotation=45)
plt.title("Average Salary by Title")
plt.xlabel("Title")
plt.ylabel("Average Salary")
plt.tight_layout()
plt.show()
```
